```python
import numpy as np
import  scipy.integrate
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('droptower_vdata.txt', unpack = True)
vel = data[1]
t = data[0]
pos =scipy.integrate.cumtrapz(vel, initial = 0)
acc = np.diff(vel)

f, (ax0, ax1,ax2) = plt.subplots(3)
ax0.scatter(t, pos)
ax0.set_title('Position vs. time')
ax1.scatter(t, vel)
ax1.set_title('Velocity vs. time')
ax2.scatter(t[:-1], acc)
ax2.set_title('Acceleration vs. time')
ax2.set_xlim([-1,11])
ax1.set_xlim([-1,11])
ax0.set_xlim([-1,11])
plt.subplots_adjust(top=1.5)
plt.gcf().set_size_inches(15,5)
plt.show()
```

```python
pos_acc = acc[acc>0]
m = np.mean(pos_acc)
print('The average positive acceleration of the ride is: ' + str(m) + ' m/s^2')
```
